```scala
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

int **pedirMatriz(int filas, int columnas){
    int **p;
    p = (int  **) malloc(filas * sizeof(int *));
    for (int i = 0; i < filas; ++i) {
        p[i] = (int *) malloc(columnas * sizeof(int));
        for (int j = 0; j < columnas; ++j) {
            p[i][j] = rand() % 11;
        }
    }return p;
}

void dibujar(int filas, int columnas,int **matrizGen){
    printf("\n");
    for (int i = 0; i < filas; ++i) {
        for (int j = 0; j < columnas; ++j) {
            printf("%i\t", matrizGen[i][j]);
        }
        printf("\n");
    }
}


int **invertir(int filas, int columnas, int **matrizGen, int **matrizInv) {
    for (int i = 0; i < filas; ++i) {
        for (int j = 0; j < columnas; ++j) {
            if (i == j){
                matrizInv[(filas - 1) - i][(columnas - 1) - j] = matrizGen[i][j];
            }
            else{
            matrizInv[j][i] = matrizGen[i][j];
        }
    }
    }
    return matrizGen;
}


void liberar(int filas, int **matrizGen){
    for (int i = 0; i<filas; i++){
        free(matrizGen[i]);
    }
    free(matrizGen);
}

void diag(int filas, int columnas, int **matrizGen){
    printf("\n");

    int tamanio = (columnas - 1);
    for (int i = 0; i < filas; ++i) {
        for (int j = 0; j < columnas; ++j) {
            if (i == j){
                printf("%i ",matrizGen[i][j]);
            }
        }
    }
    printf("\n");

    for (int i = 0; i < filas; ++i) {
        if(tamanio < 0)break;
        printf("%i ",matrizGen[i][tamanio]);
        tamanio -= 1;
    }
}



int main(){
    int **matriz, **matrizInv;
    int filas, columnas;
    srand(time(NULL));
    printf("Dame el numero de filas y columnas que quieras separados por un espacio: ");
    scanf("%i %i",&filas,&columnas);

    matriz = pedirMatriz(filas, columnas);
    dibujar(filas,columnas,matriz);
    matrizInv = pedirMatriz(filas,columnas);

    invertir(filas,columnas,matriz,matrizInv);
    dibujar(filas,columnas,matrizInv);

    diag(filas,columnas,matrizInv);
    liberar(filas,matriz);
    liberar(filas,matrizInv);
    return 0;
}
```

### 1- Pirámide:

```scala

void piramide(int n){
    if (n % 2 == 0){
        printf("El numero es impar");
        return;
    }
    int numFilas = n/2 +1;
    int max = 0;
    int min = n/2;

    for (int fila = 1; fila <= numFilas; ++fila) {
        for (int col = 1; col <= numFilas + max; ++col) {
            if (col > min) printf("*");
            else printf(" ");
        }
        printf("\n");
        max ++;
        min --;
    }
}
```

### 2-Fibonacci:

```scala

void fibonacci(int n){
    if (n<=2){
        printf("El numero debe de ser mayor que 2");
        return;
    }
    int serie[n];
    serie[0] = serie[1] = 1;
    printf("%i %i",serie[0],serie[1]);

    int i = 2;
    do {
        serie[i] = serie[i-1] + serie[i-2];
        printf("%i ",serie[i]);
        i++;
    } while (i<n);
    printf("\n");
    for (int j = n-1; j >= 0; --j) {
        printf("%i ",serie[j]);
    }
}
```

### 3-N en Raya:

```scala
// Comprueba diagonal inversa:
 #define JUGADORX 'X'
#define JUGADORO 'O'

char **creaTablero(int n) {
    char **matriz = malloc(n * sizeof (char *));
    for (int i = 0; i < n; ++i) {
        matriz[i] = malloc(n * sizeof (char));
    }

    srand(time(NULL));
    for (int i = 0; i < n; ++i) {
        for (int j = 0; j < n; ++j) {
            if (rand()%2 == 0) matriz[i][j] = JUGADORX;
            else matriz[i][j] = JUGADORO;
        }
    }

    for (int i = 0; i < n; ++i) {
        for (int j = 0; j < n; ++j) {
            printf("%c", matriz[i][j]);
        }
        printf("\n");
    }

    return matriz;
}

void comprueba(int n, char **matriz) {
    int rayasX = 0, rayasO = 0, contX = 0, contO = 0;

    // Comprueba filas:
    for (int i = 0; i < n; ++i) {
        for (int j = 0; j < n; ++j) {
            if (matriz[i][j] == JUGADORX) contX++;
            else contO++;
        }
        if (contX == n) rayasX++;
        if (contO == n) rayasO++;
        contX = contO = 0;
    }

    // Comprueba columnas:
    for (int j = 0; j < n; ++j) {
        for (int i = 0; i < n; ++i) {
            if (matriz[i][j] == JUGADORX) contX++;
            else contO++;
        }
        if (contX == n) rayasX++;
        if (contO == n) rayasO++;
        contX = contO = 0;
    }

    // Comprueba diagonal principal:
    for (int i = 0; i < n; ++i) {
        if (matriz[i][i] == JUGADORX) contX++;
        else contO++;
    }
    if (contX == n) rayasX++;
    if (contO == n) rayasO++;
    contX = contO = 0;

 
    for (int i = 0; i < n; ++i) {
        if (matriz[i][n-1-i] == JUGADORX) contX++;
        else contO++;
    }
    if (contX == n) rayasX++;
    if (contO == n) rayasO++;
    contX = contO = 0;


    printf("\nJugador X: %i", rayasX);
    printf("\nJugador O: %i", rayasO);


    for (int i = 0; i < n; ++i) {
        free(matriz[i]);
    }
    free(matriz);
}

```

# $\underline{·Ficheros: }$

Es un elemento de información persistente 
- Ruta Absoluta: empieza en el origen del sistema de ficheros.    C:\prog\fichero.txt
- Ruta Relativa: empieza en el directorio de ejecución o de trabajo.   datos\fichero.txt (adelante); \.. (atras)

·Tipos de ficheros:
- Texto: contiene caracteres ASCII u otra codificación de texto. Los entienden todos los lenguajes. Acceso secuencial
-Binario: traducir el código a binario. No son universales entre los idiomas. Acceso directo.

### $\underline{·Manejo \ de \ Ficheros \ en \ C}$

- Apertura del fichero. Asignar una variable a la apertura del fichero. FILE *f = f.open("texto.txt", "tipo de apertura").
    - "r": leer en el fichero.
    - "w": escribir en el fichero (Machaca la info existente).
    - "a": añadir al fichero, no sobrescribe la info previa. Lo abre al final y empieza a escribir allí.
    - "r+": Lectura y escritura.
    - "w+": escritura y lectura.
- Operación fscanf(), fgets()
    - Lectura: fscanf(fichero,"%s",linea), fgets(linea, tamaño, fichero), fputc()lee carácter a carácter.
    - Escritura: fprintf(fichero/stdout,"lo q quieras"), fputs(), fputc.
- Cerrar el fichero. fclose()

- Más operaciones:
    - feof() nos dice si hemos llegado al final del fichero.
    - rewind() nos vuelve a poner el puntero en el inicio del fichero. 

```scala

#include <stdio.h>
#include <string.h>
#define MAX 100


int main(){
    char linea[MAX];

    FILE *fichero = fopen("texto.txt","r"); //Apertura del fichero y lectura.
    if (fichero == NULL){
        perror("Error en apertura: "); //Función que devuelve el error específico si ocurre
        return -1;
    }

    printf("Fichero abierto OK");


    printf("\nContenido:\n------\n");

    while (!feof(fichero)){  //Mientras no sea final de fichero lo leo línea a línea
        fgets(linea, MAX ,fichero); //Leo linea a linea.
        printf("%s",linea); //Imprimo la linea.
        fscanf(fichero, "%s",linea); //Leo palabra a palbra. Se guarda seguido, el array no tiene espacio.
        printf("%s ",linea);                     //Pa eso le pongo el espacio despues del %s.
    }

    printf("------\n");


    if ( fclose(fichero)){ //Clausura del fichero.
        perror("Error en clausura de fichero: ");
        return -2;
    }
    return 0;
}
```
